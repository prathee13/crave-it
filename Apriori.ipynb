```
#Importing Libraries
import numpy as np 
import pandas as pd 
import os
import io
import pickle
import joblib
from google.colab import files
from mlxtend.frequent_patterns import apriori, association_rules 
from mlxtend.preprocessing import TransactionEncoder
```

```
uploaded = files.upload()
```

```
data1 = pd.read_excel(io.BytesIO(uploaded['store_data.xlsx']), header=None)
data2 = data1.to_numpy() 
data3 = []

for i in data2:
  new_row = []
  for j in i:
    if j == j and j != '                       ':
      new_row.append(j)
  data3.append(np.array(new_row))

# data3 = np.array(data3)
data3 = np.array(data3)  
data3
```

```
te = TransactionEncoder()
te_ary = te.fit(data3).transform(data3)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```
filename = 'dogs'
outfile = open(filename,'wb')

#saving trained model to Pickle
saved_model = pickle.dump(frequent_itemsets, outfile)
outfile.close()


infile = open(filename,'rb')
new_dict = pickle.load(infile)
infile.close()
#Load the pickled model
# apriori_pickled = pickle.loads(saved_model)
new_dict
#Use the loaded model 
#same as frequent_itemsets[ (frequent_itemsets['length'] >= 2)]
#change the length as you wish. 
# apriori_pickled[ (apriori_pickled['length'] >= 3)]
# apriori_pickled
```

```
# Save the model as a pickle in a file 
#The pickled file can be found in the 'Table of contents' or also in Github
joblib.dump(frequent_itemsets, 'apriori.pkl') 

# Load the model from the file 
apriori_joblib = joblib.load('apriori.pkl')

# Use the loaded model to make predictions 
apriori_joblib[(apriori_joblib ['length'] >= 3)]
```
